```python
mAP = [0.766, 0.769, 0.785, 0.787, 0.793]
nano_pytorch_ms = [160.65, 288.14, 653.34, 1070.74]
nano_onnx_ms = [133.30, 217.20, 471.59, 733.27, 1208.69]
cpu_ms = [353.13, 891.93, 1792.31, 3340.26, 4222.87]
gpu_onnx_ms = [12.04, 12.16, 20.20, 29.09, 41.87]

import numpy as np
nano_pytorch_fps = list(1. / np.array(nano_pytorch_ms))
nano_onnx_fps = list(1. / np.array(nano_onnx_ms))
cpu_fps = list(1. / np.array(cpu_ms))
# a100_fps = list(1. / np.array(a100_ms))
```
